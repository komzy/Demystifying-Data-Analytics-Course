# **5b. Data Transformation & Manipulation**
By data transformation and manipulation we mean to change the data representation in order to support certain analytical tasks. Normalization is one form of transformation which can be of different types as shown below:
Normalization.jpg

```python
#a. Normalization
confirmed_cases_zscore = confirmed_cases.iloc[:, 4:497]
confirmed_cases_zscore
for column in confirmed_cases_zscore.columns:
 confirmed_cases_zscore[column] = (confirmed_cases_zscore[column] -
                          confirmed_cases_zscore[column].mean()) / confirmed_cases_zscore[column].std()
display(confirmed_cases_zscore)

confirmed_cases_min_max = confirmed_cases.iloc[:, 4:496]
for column in confirmed_cases_min_max.columns:
    confirmed_cases_min_max[column] = (confirmed_cases_min_max[column] - confirmed_cases_min_max[column].min()) / (confirmed_cases_min_max[column].max() - confirmed_cases_min_max[column].min())  
display(confirmed_cases_min_max)
```

```python
#b. Country-wise confirmed, active, recovered and death cases
unique_countries = list(confirmed_cases['Country/Region'].unique())
print(len(unique_countries))

country_wise_stats = pd.DataFrame(columns = ['Country', 'Confirmed', 'Active', 'Recovered', 'Death', 'Mortality Rate', 'Recovery Rate'])
print(country_wise_stats)

for i in unique_countries:
  confirmed_value = confirmed_cases[confirmed_cases['Country/Region']==i].iloc[:, 4:497].sum(axis = 1, skipna = True).sum()
  recovered_value = recovered_cases[recovered_cases['Country/Region']==i].iloc[:, 4:497].sum(axis = 1, skipna = True).sum()
  death_value = death_cases[death_cases['Country/Region']==i].iloc[:, 4:497].sum(axis = 1, skipna = True).sum()
  active_value = confirmed_value-recovered_value-death_value
  mortality_rate = death_value/confirmed_value
  recovery_rate = recovered_value/confirmed_value
  to_append = [i, confirmed_value, active_value, recovered_value, death_value, mortality_rate, recovery_rate ]
  country_wise_stats.loc[len(country_wise_stats)] = to_append

country_wise_stats
```

```python
#c. Performing Statistical Analysis through 5-numbered summary 
from numpy import percentile
# calculate quartiles
quartiles = percentile(country_wise_stats['Confirmed'], [25, 50, 75])
# calculate min/max
data_min, data_max = country_wise_stats['Confirmed'].min(), country_wise_stats['Confirmed'].max()
# print 5-number summary
print('Min:', data_min)
print('Q1', quartiles[0])
print('Median:', quartiles[1])
print('Q3:',  quartiles[2])
print('Max:', data_max)
```

```python
#d. Adding a new column to a dataframe
country_wise_stats["Intensity"] = " "

country_wise_stats.loc[(country_wise_stats['Confirmed'] >= data_min) & (country_wise_stats['Confirmed'] < quartiles[0]), 'Intensity'] = 'Low'
country_wise_stats.loc[(country_wise_stats['Confirmed'] >= quartiles[0]) & (country_wise_stats['Confirmed'] < quartiles[1]), 'Intensity'] = 'Medium'
country_wise_stats.loc[(country_wise_stats['Confirmed'] >= quartiles[1]) & (country_wise_stats['Confirmed'] < quartiles[2]), 'Intensity'] = 'High'
country_wise_stats.loc[(country_wise_stats['Confirmed'] >= quartiles[2]) & (country_wise_stats['Confirmed'] <= data_max), 'Intensity'] = 'Critical'
country_wise_stats
```

```python
#e. Changing categorical data into numeric-Using Renaming and Dropping (DEBUGGED)
country_wise_stats['Codes'] = pd.factorize(country_wise_stats.Intensity)[0] #Error was: In country_wise_stats dataframe a new column needs to be added that contains the encoded arraty 
country_wise_stats = country_wise_stats.drop(['Intensity'], axis=1) #Then drop the old categorical column
country_wise_stats.rename(columns={'Codes':'Intensity'}) # Then rename the new column
```

```python
country_wise_stats.dtypes
```

```python
#f. Changing data type from into object to numeric
print(country_wise_stats.dtypes)
country_wise_stats["Confirmed"] = country_wise_stats["Confirmed"].astype(object).astype(int)
country_wise_stats["Active"] = country_wise_stats["Active"].astype(object).astype(int)
country_wise_stats["Death"] = country_wise_stats["Death"].astype(object).astype(int)
country_wise_stats["Recovered"] = country_wise_stats["Recovered"].astype(object).astype(int)
print(country_wise_stats.dtypes)
```

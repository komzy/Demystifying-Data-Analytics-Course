**Import Death, Recovered and Latest Dataset from John Hopkins Repository**

```python
import numpy as np
import pandas as pd

#deaths
deaths_global=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
deaths_global.head()
```

```python
#recovered
recovered_global=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
recovered_global.head()
```

```python
#latest (26th-May-2021 at the time of creation of this assignment)
latest=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/05-26-2021.csv')
latest.head()
```

**Exploration with Pandas**

```python
type(recovered_global)
```

```python
daily_recovered_=recovered_global.drop(['Province/State', 'Lat','Long'], axis=1)
daily_recovered_=daily_recovered_.groupby("Country/Region").sum()
daily_recovered=daily_recovered_.transpose()
daily_recovered.head()
```

```python
daily_recovered.plot(figsize=(15,10)).legend(loc='left',bbox_to_anchor=(1.0, 1.0))
```

```python
daily_deaths_=deaths_global.drop(['Province/State', 'Lat','Long'], axis=1)
daily_deaths_=daily_deaths_.groupby("Country/Region").sum()
daily_deaths=daily_deaths_.transpose()
daily_deaths.head()
```

```python
daily_deaths.plot(figsize=(15,10)).legend(loc='left',bbox_to_anchor=(1.0, 1.0))
```

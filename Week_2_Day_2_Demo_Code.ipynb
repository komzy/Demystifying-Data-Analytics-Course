```python
#Importing the required libraries
import pandas as pd
import numpy as np

# Fetching the COVID-19 confirmed cases data from John Hopkins repository
confirmed_cases = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv")

confirmed_month_wise = pd.DataFrame(columns = ['Country', 'Jan-20', 'Feb-20', 'Mar-20', 'Apr-20', 'May-20', 'Jun-20', 'Jul-20', 'Aug-20', 'Sep-20', 'Oct-20', 'Nov-20', 'Dec-20', 'Jan-21', 'Feb-21', 'Mar-21', 'Apr-21', 'May-21'])

unique_countries = list(confirmed_cases['Country/Region'].unique())
print(len(unique_countries))

#Using mean instead of sum
china_jan_20 = confirmed_cases[confirmed_cases['Country/Region']=='China'].loc[:, '1/22/20':'1/31/20'].sum(axis = 1, skipna = True).sum()
print(china_jan_20)

for i in unique_countries:
  jan_20 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '1/22/20':'1/31/20'].sum(axis = 1, skipna = True).sum()
  feb_20 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '2/1/20':'2/28/20'].sum(axis = 1, skipna = True).sum()
  mar_20 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '3/1/20':'3/31/20'].sum(axis = 1, skipna = True).sum()
  apr_20 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '4/1/20':'4/30/20'].sum(axis = 1, skipna = True).sum()
  may_20 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '5/1/20':'5/31/20'].sum(axis = 1, skipna = True).sum()
  jun_20 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '6/1/20':'6/30/20'].sum(axis = 1, skipna = True).sum()
  jul_20 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '7/1/20':'7/31/20'].sum(axis = 1, skipna = True).sum()
  aug_20 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '8/1/20':'8/31/20'].sum(axis = 1, skipna = True).sum()
  sep_20 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '9/1/20':'9/30/20'].sum(axis = 1, skipna = True).sum()
  oct_20 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '10/1/20':'10/31/20'].sum(axis = 1, skipna = True).sum()
  nov_20 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '11/1/20':'11/30/20'].sum(axis = 1, skipna = True).sum()
  dec_20 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '12/1/20':'12/31/20'].sum(axis = 1, skipna = True).sum()
  jan_21 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '1/1/21':'1/31/21'].sum(axis = 1, skipna = True).sum()
  feb_21 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '2/1/21':'2/28/21'].sum(axis = 1, skipna = True).sum()
  mar_21 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '3/1/21':'3/31/21'].sum(axis = 1, skipna = True).sum()
  apr_21 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '4/1/21':'4/30/21'].sum(axis = 1, skipna = True).sum()
  may_21 = confirmed_cases[confirmed_cases['Country/Region']==i].loc[:, '5/1/21':'5/28/21'].sum(axis = 1, skipna = True).sum()
  to_append = [i, jan_20, feb_20, mar_20, apr_20, may_20, jun_20, jul_20, aug_20, sep_20, oct_20, nov_20, dec_20, jan_21, feb_21, mar_21, apr_21, may_21]
  confirmed_month_wise.loc[len(confirmed_month_wise)] = to_append

print(confirmed_month_wise)
```

```python
#Importing the required libraries
from matplotlib import pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
#a. Generating the correlation matrix
corr_matrix = confirmed_month_wise.corr(method='pearson')
print(corr_matrix)
plt.matshow(corr_matrix, cmap="autumn") #Display an array as a matrix in a new figure window.
plt.colorbar()

plt.xticks(list(range(len(corr_matrix.columns))), corr_matrix.columns, rotation='vertical')
plt.yticks(list(range(len(corr_matrix.columns))), corr_matrix.columns, rotation='horizontal')

plt.show()
```

```python
#b.Boxplot based on 5-numbered summary
#corr_matrix['May-21'].sort_values(ascending=False)
corr_matrix.plot.box() # A dataframe has its own limited set of functions for plotting as well
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
#c. Plotting single column or a complete dataframe
plt.plot(confirmed_month_wise['May-21'].values)
plt.xlabel('Number of Values/Index')
plt.ylabel('Range of Values')
plt.title('Column-level Plot')
plt.show()

confirmed_month_wise.plot()
plt.xlabel('Number of Values/Index')
plt.ylabel('Range of Values')
plt.title('Dataframe-level Plot')
plt.show()
```

```python
#d. Plotting the country-wise yearly summary in terms of STACKED PLOT-- Showing number of confirmed cases in each quarter
x = ['Pakistan', 'India', 'United Kingdom', 'China', 'France'] #Let's compare Pakistan with other countries
y1 = pd.Series()
y2 = pd.Series()
y3 = pd.Series()
y4 = pd.Series()
y5 = pd.Series()

#Getting quarterly data for y-axis
for i in x:
  value = confirmed_month_wise[confirmed_month_wise['Country']==i].loc[:,'Apr-21']
  y1 = y1.append(value) 
for i in x:
  value = confirmed_month_wise[confirmed_month_wise['Country']==i].loc[:,'Jan-21']
  y2 = y2.append(value) 
for i in x:
  value = confirmed_month_wise[confirmed_month_wise['Country']==i].loc[:,'Sep-20']
  y3 = y3.append(value) 
for i in x:
  value = confirmed_month_wise[confirmed_month_wise['Country']==i].loc[:,'Jul-20']
  y4 = y4.append(value) 
for i in x:
  value = confirmed_month_wise[confirmed_month_wise['Country']==i].loc[:,'Apr-20']
  y5 = y5.append(value) 

plt.bar(x, y1, color='g')
plt.bar(x, y2, color='y')
plt.bar(x, y3, color='b')
plt.bar(x, y4, color='k')
plt.bar(x, y5, color='r')

plt.xlabel('Countries')
plt.ylabel('Apr21-Apr20')
plt.title('Stacked Plot')
plt.show()

```

```python
#f. Demostrating SWARMPLOT

# Fetching some new data
recovered_cases = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
death_cases = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
from numpy import percentile
unique_countries = list(confirmed_cases['Country/Region'].unique())
country_wise_stats = pd.DataFrame(columns = ['Country', 'Confirmed', 'Active', 'Recovered', 'Death', 'Mortality Rate', 'Recovery Rate'])

for i in unique_countries:
  confirmed_value = confirmed_cases[confirmed_cases['Country/Region']==i].iloc[:, 4:497].sum(axis = 1, skipna = True).sum()
  recovered_value = recovered_cases[recovered_cases['Country/Region']==i].iloc[:, 4:497].sum(axis = 1, skipna = True).sum()
  death_value = death_cases[death_cases['Country/Region']==i].iloc[:, 4:497].sum(axis = 1, skipna = True).sum()
  active_value = confirmed_value-recovered_value-death_value
  mortality_rate = death_value/confirmed_value
  recovery_rate = recovered_value/confirmed_value
  to_append = [i, confirmed_value, active_value, recovered_value, death_value, mortality_rate, recovery_rate ]
  country_wise_stats.loc[len(country_wise_stats)] = to_append

x = ['Pakistan', 'India', 'United Kingdom', 'China', 'France'] #Let's compare Pakistan with other countries = pd.Series()
y1 = pd.Series() # Confirmed
y2 = pd.Series() # Recovered
y3 = pd.Series() #Active

#Getting quarterly data for y-axis
for i in x:
  value = country_wise_stats[country_wise_stats['Country']==i].loc[:,'Confirmed']
  y1 = y1.append(value) 
for i in x:
  value = country_wise_stats[country_wise_stats['Country']==i].loc[:,'Recovered']
  y2 = y2.append(value) 
for i in x:
  value = country_wise_stats[country_wise_stats['Country']==i].loc[:,'Active']
  y3 = y3.append(value) 

new_frame = { 'Country': x, 'Confirmed': y1, 'Recovered': y2, 'Active': y3  }
result = pd.DataFrame(new_frame)

# Swarmplot
sns.swarmplot(x='Country', y='Confirmed', data=result)
plt.show()
```

```python
#h. Generating country-wise GRADIENT MAP
import operator #Like mathematical operator, here we are using for sorting
latest_data = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/05-29-2021.csv')
unique_countries = list(latest_data['Country_Region'].unique())
unique_countries
country_confirmed_cases = []
country_death_cases = []
country_active_cases = []
country_recovery_cases = []
country_mortality_rate = []
country_recovery_rate = []
no_cases = []

for i in unique_countries:
  cases = latest_data[latest_data['Country_Region']==i]['Confirmed'].sum()
  if cases > 0:
    country_confirmed_cases.append(cases)
  else:
    no_cases.append(i) 

for i in no_cases:
    unique_countries.remove(i)

#sort the countries according to confirmed cases
unique_countries = [k for k, v in sorted (zip(unique_countries, country_confirmed_cases), key=operator.itemgetter(1), reverse=True)]
for i in range(len(unique_countries)):
  country_confirmed_cases[i] = latest_data[latest_data['Country_Region']==unique_countries[i]]['Confirmed'].sum()
  country_death_cases.append(latest_data[latest_data['Country_Region']==unique_countries[i]]['Deaths'].sum())
  country_recovery_cases.append(latest_data[latest_data['Country_Region']==unique_countries[i]]['Recovered'].sum())
  country_active_cases.append(country_confirmed_cases[i]-country_death_cases[i]-country_recovery_cases[i])
  country_mortality_rate.append(country_death_cases[i]/country_confirmed_cases[i])
  country_recovery_rate.append(country_recovery_cases[i]/country_confirmed_cases[i])

country_df =pd.DataFrame({'Country Name': unique_countries, 'Number of Confirmed Cases': country_confirmed_cases, 'Number of Deaths': country_death_cases, 'Number of Active Cases': country_active_cases, 'Number of Recovered Cases': country_recovery_cases, 'Mortality Rate': country_mortality_rate, 'Recovery Rate':country_recovery_rate})
country_df.style.background_gradient(cmap='Blues') #This is a very simpler version of visualization of the intensity of the values within a dataframe by styling it in gradient form.
```
